## KDE

```python
kde = stats.gaussian_kde(aapl.r.dropna())
```

```python
new_data = kde.resample(5000)
```

```python
new_data = new_data.reshape((5000,))
```

```python
plt.hist(new_data, bins=50, range=(-0.2,0.2));
```
